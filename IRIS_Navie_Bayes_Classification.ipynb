```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('IRIS Navie Bayes Classification')\
                    .getOrCreate()
```

```python
spark
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
dataInstance = load_iris()
dataInstance
```

```python
data = dataInstance.data
data
```

```python
dataframe = pd.DataFrame(data=data, columns=dataInstance.feature_names)
dataframe
```

```python
dataframe['target'] = dataInstance.target
dataframe
```

```python
sparkDataframe = spark.createDataFrame(dataframe)
```

```python
sparkDataframe.show()
```

```python
sparkDataframe.printSchema()
```

```python
sparkDataframe.columns
```

```python
from pyspark.ml.feature import VectorAssembler
```

```python
featureAssembler = VectorAssembler(inputCols=['sepal length (cm)',
 'sepal width (cm)',
 'petal length (cm)',
 'petal width (cm)'], outputCol='features')
```

```python
sparkDataframe = featureAssembler.transform(sparkDataframe)
```

```python
sparkDataframe.show()
```

```python
modelData = sparkDataframe.select(['features', 'target'])
```

```python
modelData.show()
```

```python
trainData, testData = modelData.randomSplit([0.8, 0.2], seed = 2345)
```

```python
trainData.show()
```

```python
testData.show()
```

```python
from pyspark.ml.classification import NaiveBayes
```

```python
navieBayesModel = NaiveBayes(featuresCol='features', labelCol='target')
```

```python
navieBayesModel = navieBayesModel.fit(trainData)
```

```python
prediction = navieBayesModel.transform(testData)
```

```python
prediction.show()
```

```python
prediction.groupBy(['target', 'prediction']).count().show()
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```python
evaluator = MulticlassClassificationEvaluator(labelCol='target', predictionCol='prediction')
```

```python
accuracy = evaluator.evaluate(prediction)
accuracy
```

```python
spark.stop()
```
